### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Data

```python
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

### Get information about the dataset


```python
df.info()
```

### Check for missing values

```python
df.isnull().sum()
```

### To check the total number of duplicated rows


```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

### Scatter plot of Profit vs Sales

```python
sns.scatterplot(x='Sales', y='Profit', data=df)
```

### To view correlation coefficients between all values

```python
corr_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

### To view the relationship between 'Category' and 'Profit'

```python
sns.barplot(x='Category', y='Profit', data=df) 
```

### Calculate the total sales and profit by region

```python
region_data = df.groupby('Region').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
region_data
```

### Group the data by region and calculate the total sales and profit

```python
sns.set_style('whitegrid')
sns.barplot(x='Region', y='Sales', data=region_data, color='lightblue', label='Sales')
sns.barplot(x='Region', y='Profit', data=region_data, color='lightgreen', label='Profit')

plt.title('Sales and Profit by Region')
plt.xlabel('Region')
plt.ylabel('Amount')
plt.legend()

plt.show()
```
